# **Simple Linear Regresion on Car Prediction**

Dataset: https://www.kaggle.com/datasets/hellbuoy/car-price-prediction

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
car_df = pd.read_csv('/content/CarPrice_Assignment.csv')
```

```python
car_df.columns
```

```python
car_df.iloc[0]
```

```python
label_encoder = LabelEncoder()

car_df['CarName'] = label_encoder.fit_transform(car_df['CarName'])
car_df['fueltype'] = label_encoder.fit_transform(car_df['fueltype'])
car_df['aspiration'] = label_encoder.fit_transform(car_df['aspiration'])
car_df['doornumber'] = label_encoder.fit_transform(car_df['doornumber'])
car_df['carbody'] = label_encoder.fit_transform(car_df['carbody'])
car_df['drivewheel'] = label_encoder.fit_transform(car_df['drivewheel'])
car_df['enginelocation'] = label_encoder.fit_transform(car_df['enginelocation'])
car_df['enginetype'] = label_encoder.fit_transform(car_df['enginetype'])
car_df['cylindernumber'] = label_encoder.fit_transform(car_df['cylindernumber'])
car_df['fuelsystem'] = label_encoder.fit_transform(car_df['fuelsystem'])


car_df.head(5)
```

```python
y = car_df['price']
```

```python
X = car_df.drop(['price', 'car_ID'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
model.score(X_test,y_test)
```

```python
# Step 8 : model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
