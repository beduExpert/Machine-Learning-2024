# **Simple Linear Regresion on Insurance Data**

Dataset: https://www.kaggle.com/datasets/willianoliveiragibin/healthcare-insurance

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
insurance = pd.read_csv('/content/insurance.csv')
```

```python
insurance.columns
```

```python
label_encoder = LabelEncoder()

insurance['sex'] = label_encoder.fit_transform(insurance['sex'])
insurance['smoker'] = label_encoder.fit_transform(insurance['smoker'])
insurance['region'] = label_encoder.fit_transform(insurance['region'])
```

```python
insurance
```

```python
y = insurance['charges']
```

```python
X = insurance.drop(['charges'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Step 8 : model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
